```python
import pandas as pd
import json
```

```python
city = 'chicago'
df = pd.read_csv(f'./data/{city}_subzone_attr.csv')
colnames = ['Planning Area'] + list(df.columns[1:])
df.columns = colnames
```

```python
results = df.to_json(orient='records')
```

```python
parsed = json.loads(results)
```

```python
with open('../data/bangkok.json', 'w') as outfile:
    json.dump(parsed, outfile)
```
